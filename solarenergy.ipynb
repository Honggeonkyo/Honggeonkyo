```python
# 전처리에서 normalization을 진행했다. 전데이터 전부다 0~1까지로 바꾸어 주었다.
from sklearn import preprocessing
#x=solar.values.astype(np.float32)
normal=preprocessing.MinMaxScaler()
x_scaled=normal.fit_transform(solar)
solar=pd.DataFrame(x_scaled)
solar=solar.rename(columns={0:'정격용량',1:'년',2:'월',3:'일',4:'시',5:'인버터출력',6:'경사일사량',7:'수평일사량',8:'표면온도',9:'주위온도',10:'y'})
solar[['인버터출력','주위온도']]
```

```python
X=solar[['인버터출력','주위온도']]
y=solar['y']
x1=solar['인버터출력']
x2=solar['주위온도']
```

```python
plt.scatter(x1[y==1],x2[y==1],label='정상')
plt.scatter(x1[y==0],x2[y==0],label='고장')
plt.rcParams["figure.figsize"]=(10,100)
plt.title('broken and normal')
plt.xlabel("x1")
plt.ylabel('x2')
```

```python
#'gamma': [1, 0.1, 0.01, 0.001, 0.00001, 10]
#gamma=0.1
from sklearn import svm
from sklearn.model_selection import GridSearchCV
clf = svm.SVC(kernel='linear', C = 10.0)

# Grid Search
# Parameter Grid
param_grid = {'C': [0.1, 1, 10, 100]}
 
# Make grid search classifier
clf_grid = GridSearchCV(svm.SVC(), param_grid, verbose=1)
 
# Train the classifier
clf_grid.fit(X, y)
 
# clf = grid.best_estimator_()
print("Best Parameters:\n", clf_grid.best_params_)
print("Best Estimators:\n", clf_grid.best_estimator_)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import numpy as np
h_class=SVC(kernel='linear',C=10)
scores=cross_val_score(h_class,X,y,cv=3,scoring='accuracy')
print('Accuracy: %0.3f' %np.mean(scores))
```

```python
h_class.fit(X.values,y.values)
print(h_class.support_)
```

```python
y=y.values.astype(np.int64)
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X=X.values,y=y.ravel(),clf=h_class,legend=1)
#plt.xlabel(x1.columns,size=14)
#plt.ylabel(x2.columns,size=14)
plt.title('SVM Decision Region Boundary',size=16)
plt.rcParams["figure.figsize"]=(10,100)
```
